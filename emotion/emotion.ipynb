```python
import cv2
from deepface import DeepFace


# Load Haar-Cascade untuk deteksi wajah
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


# Buka kamera eksternal
cap = cv2.VideoCapture(1)  # Coba dengan indeks yang sesuai untuk kamera eksternal

if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Jika gagal, coba dengan indeks lain


def analyze_emotion(face_img):
    """
    Analisis ekspresi wajah dari wajah yang sudah terdeteksi menggunakan DeepFace.
    """
    try:
        # Analisis ekspresi wajah dengan DeepFace
        result = DeepFace.analyze(face_img, actions=["emotion"], enforce_detection=False, silent=True)
        # Ambil ekspresi dominan dari hasil analisis
        dominant_emotion = result[0]['dominant_emotion']
        return dominant_emotion
    except Exception as e:
        print(f"Error dalam analisis ekspresi: {e}")
        return "Unknown"


while True:
    # Tangkap frame dari kamera
    ret, frame = cap.read()
    if not ret:
        print("Gagal mengambil frame")
        break

    # Konversi gambar ke grayscale untuk deteksi yang lebih cepat dan akurat
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Deteksi wajah dengan Haar-Cascade
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.3, minSize=(30, 30))

    # Gambar kotak di sekitar wajah yang terdeteksi
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # Potong wajah yang terdeteksi
        face_img = frame[y:y + h, x:x + w]

        # Analisis ekspresi wajah
        emotion = analyze_emotion(face_img)

        # Tampilkan teks analisis ekspresi di atas kotak wajah
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, emotion, (x, y - 10), font, 0.9, (255, 0, 0), 2)

    # Tampilkan hasil deteksi wajah dan analisis ekspresi
    cv2.imshow('Deteksi dan Analisis Ekspresi', frame)

    # Tekan 'q' untuk keluar
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Lepaskan sumber daya
cap.release()
cv2.destroyAllWindows()
```
